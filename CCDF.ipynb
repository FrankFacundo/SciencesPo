```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

```python
frequencies = {"A":10, "B":2,"C":3, "D":4, "E":15,"F":2,"G":4,
               "H":2,"I":12,"J":3,"K":2,"L":6,"M":3,"N":8,"O":10,
               "P":3,"Q":2,"R":8,"S":6,"T":8,"U":6,"V":3,"W":2,
               "X":1,"Y":2,"Z":1,
              }
```

```python
fig = plt.figure()
plt.bar(range(len(frequencies)), list(frequencies.values()), 
        align='center')
plt.xticks(range(len(frequencies)), list(frequencies.keys()))

xlocs, xlabs = plt.xticks()
xlocs=[i for i in range(0,len(frequencies))]
xlabs=[i/2 for i in range(0,len(frequencies))]
for i, v in enumerate(list(frequencies.values())):
    plt.text(xlocs[i]-0.25, v + 0.2, str(v))

plt.title("Frequency")
#plt.show()
#fig.savefig('Frequency.png')
```

```python
fig = plt.figure()
total = sum(frequencies.values())
pdf = np.array(list(frequencies.values()))/total
plt.bar(range(len(frequencies)), pdf, align='center')
plt.xticks(range(len(frequencies)), list(frequencies.keys()))

xlocs, xlabs = plt.xticks()
xlocs=[i for i in range(0,len(frequencies))]
xlabs=[i/2 for i in range(0,len(frequencies))]
for i, v in enumerate(pdf):
    plt.text(xlocs[i] - 0.5, v+0.001, str("{0:.3f}".format(v)))

plt.title("Probability Mass Function (PDF)")
#plt.show()
#fig.savefig('PDF.png')
```

```python
cdf = np.cumsum(pdf)
```

```python
fig = plt.figure()
plt.plot(range(len(frequencies)),cdf)
plt.xticks(range(len(frequencies)), list(frequencies.keys()))

xlocs, xlabs = plt.xticks()
xlocs=[i for i in range(0,len(frequencies))]
xlabs=[i/2 for i in range(0,len(frequencies))]
for i, v in enumerate(cdf):
    plt.text(xlocs[i] - 0.5, v+0.02, str("{0:.2f}".format(v)))

plt.grid()
plt.title("Cumulative Distribution Function (CDF)")
#plt.show()
fig.savefig('CDF.png')
```

```python
fig = plt.figure()
ccdf = 1-cdf
plt.plot(range(len(frequencies)),ccdf)
plt.xticks(range(len(frequencies)), list(frequencies.keys()))

xlocs, xlabs = plt.xticks()
xlocs=[i for i in range(0,len(frequencies))]
xlabs=[i/2 for i in range(0,len(frequencies))]
for i, v in enumerate(ccdf):
    plt.text(xlocs[i] - 0.5, v+0.02, str("{0:.2f}".format(v)))

plt.grid()
plt.title("Complemetary Cumulative Distribution Function (CCDF = 1 - CDF)")
#plt.show()
fig.savefig('CCDF.png')
```
